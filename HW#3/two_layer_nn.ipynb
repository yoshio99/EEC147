## Classify CIFAR-10

Do classification on the CIFAR-10 dataset.

```python
from cs231n.data_utils import load_CIFAR10

def get_CIFAR10_data(num_training=49000, num_validation=1000, num_test=1000):
    """
    Load the CIFAR-10 dataset from disk and perform preprocessing to prepare
    it for the two-layer neural net classifier. These are the same steps as
    we used for the SVM, but condensed to a single function.  
    """
    # Load the raw CIFAR-10 data
    cifar10_dir = './cifar-10-batches-py'
    X_train, y_train, X_test, y_test = load_CIFAR10(cifar10_dir)
        
    # Subsample the data
    mask = list(range(num_training, num_training + num_validation))
    X_val = X_train[mask]
    y_val = y_train[mask]
    mask = list(range(num_training))
    X_train = X_train[mask]
    y_train = y_train[mask]
    mask = list(range(num_test))
    X_test = X_test[mask]
    y_test = y_test[mask]

    # Normalize the data: subtract the mean image
    mean_image = np.mean(X_train, axis=0)
    X_train -= mean_image
    X_val -= mean_image
    X_test -= mean_image

    # Reshape data to rows
    X_train = X_train.reshape(num_training, -1)
    X_val = X_val.reshape(num_validation, -1)
    X_test = X_test.reshape(num_test, -1)

    return X_train, y_train, X_val, y_val, X_test, y_test


# Invoke the above function to get our data.
X_train, y_train, X_val, y_val, X_test, y_test = get_CIFAR10_data()
print('Train data shape: ', X_train.shape)
print('Train labels shape: ', y_train.shape)
print('Validation data shape: ', X_val.shape)
print('Validation labels shape: ', y_val.shape)
print('Test data shape: ', X_test.shape)
print('Test labels shape: ', y_test.shape)
```

## Answers:

(1) Some reasons why the training accuracy is not good might be poor intialization of the weights and biases, the lack of normalization during forward and backward passes, and/or a poor learning rate. Since the weights were intialized to small random values and the biases to zero, which empirically cause the weights to go to zero as more iterations are performed, which leads eventually to no more learning regardless of number of iterations. Secondly, having no batch normalization results in the gradients of the weights and biases not considering that other weights and biases change in the same step, hurting the rate at which the NN can learn. Lastly, a poor learning rate is causing the loss to jump around as it gets closer to a local optimum.

(2) We can initialize the weights and biases in a better way (like with a Xavier initialization) to make sure that the gradients don't drop to zero so quickly. Also, we can add batch normalization to each of the layers to ensure that learn gradient step is the best for learning at that point. Lastly, we can tune the learning rate to find out which value is best.
